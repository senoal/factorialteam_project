# Data

```python
data_test = pd.read_csv("application_test.csv")
print("info data : ",data_test.info())
print("tipe data : ",data_test.dtypes.unique())
print("dimensi data : ",data_test.shape)
```

# Pisahkan kolom berdasarkan tipe data

```python
df1 = data_test.select_dtypes(include=['int64'])
df1.shape
```

```python
df2 = data_test.select_dtypes(include=['O'])
# df2.head(1)
# df2.columns
df2.shape
```

```python
df3 = data_test.select_dtypes(include=['float64'])
# df3.head(1)
# df3.columns
df3.shape
```

# Periksa jumlah missing values

```python
jml1 = df1.isnull().sum()/df1.shape[0]*100
jml1
```

```python
df11 = df1.isnull().sum()/df1.shape[0]*100
df11
```

```python
# cek persentase missing value
jml2 = df2.isnull().sum()/df2.shape[0]*100
jml2
```

```python
# filter columns berdasarkan jumlah missing values yang ditentukan (<60%)
jml = 60
df2 = df2.loc[:, (df2.isnull().sum()/df2.shape[0]*100 < jml)]
# df2.head(1)
df2.isnull().sum()
```

```python
# cek persentase missing value
jml3 = df3.isnull().sum()/df3.shape[0]*100
jml3.shape
```

```python
# filter columns berdasarkan jumlah missing values yang ditentukan (<60%)
jml = 60
df3 = df3.loc[:, (df3.isnull().sum()/df3.shape[0]*100 < jml)]
# df3.head(1)
# df3.columns
# df3.isnull().sum()
df3.shape
```

# Imputasi df2

```python
df2.tail()
```

```python
# munculkan mode
df2.mode()
```

```python
for column in df2:
    df2[column].fillna(df2[column].mode()[0], inplace=True)

df2.tail()
```

# Imputasi df3

```python
df3.tail()
```

```python
df3.mean()
```

```python
for i in df3:
    df3[i].fillna(df3[i].mean(), inplace=True)
df3.tail()
```
